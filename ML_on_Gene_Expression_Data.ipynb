#Basics of ML workflow (data → preprocessing → modeling)
#Step 2: **Preprocessing**

```python
# Import label encoder
from sklearn import preprocessing
#Here, we are importing the preprocessing module from scikit-learn (sklearn).
#This module contains various tools for preparing data, one of which is the LabelEncoder,
#used for converting categorical labels into numbers (needed for ML models).


# label_encoder object knows how to understand word labels.
label_encoder = preprocessing.LabelEncoder()
# we are creating an instance/object of the LabelEncoder class and store it in the variable label_encoder.
# This object has methods (fit, transform, fit_transform) that can map categorical (text) labels into numeric values.
# Example: ["Apple", "Banana", "Orange"] → [0, 1, 2].


# Encode labels in column 'label'.
combined_PRAD['labels']= label_encoder.fit_transform(combined_PRAD['labels'])

# combined_PRAD is your pandas DataFrame.
# combined_PRAD['labels'] selects the column named labels.
# fit_transform() does two things at once:
# fit → Learns all the unique categories (like Apple, Banana, Orange).
# transform → Assigns a numeric code to each category (e.g., Apple=0, Banana=1, Orange=2).
# Then, it replaces the original labels column with these numeric codes.
# So now, instead of text labels, you have integer labels in that column.

#To have a list of unique enteries.
combined_PRAD['labels'].unique()
# .unique() returns the array of unique values present in the labels column.
# This shows which numeric codes were assigned after encoding.
# Example: Output might be → [0, 1, 2].

#counting the number of classes
combined_PRAD["labels"].value_counts()
```

```python
#Assigning the numerical data to a "X" variable and labels column into a "y" variable that will be used in the next steps
X = combined_PRAD.iloc[:,:-1]
y = combined_PRAD["labels"]



# .iloc → This is pandas’ integer-location based indexing. It allows you to select rows and columns by their numeric positions (not labels).
# The first position (:) → Refers to rows.
# A plain : means all rows are selected.
# The second position (:-1) → Refers to columns.
# :-1 means from the first column up to (but not including) the last column.
# In other words, it selects all columns except the last one.
```

#Basics of ML workflow (data → preprocessing → modeling)
#Step 3: **Modeling**

```python
#importing train_test_split
from sklearn.model_selection import train_test_split
X_train, X_test ,Y_train, Y_test = train_test_split(X,y,test_size =0.30, random_state=42)
# X = features (independent variables, input data).
# y = labels/targets (dependent variable).
# train_test_split splits X and y into training and testing sets.
# 🔹 Parameters:
# test_size=0.30 → 30% of the data will be reserved for testing, 70% for training.
# random_state=42 → ensures the split is reproducible. (If you run it again with the same random_state, you’ll get the same split every time.)
# 🔹 Outputs:
# X_train → features for training set (70%).
# X_test → features for testing set (30%).
# Y_train → labels for training set (70%).
# Y_test → labels for testing set (30%).

from sklearn.preprocessing import StandardScaler
sc= StandardScaler()

# StandardScaler standardizes your features → it makes sure they all have:
# Mean = 0
# Standard deviation = 1
# Why? → Because ML models (like Logistic Regression, SVM, Neural Nets) perform better when features are on a similar scale.

X_train = sc.fit_transform(X_train)
# fit_transform does two steps on training data only:
# fit() → Learns the mean and standard deviation of each feature in X_train.
# transform() → Uses those values to scale the training data.

X_test = sc.transform(X_test)

# Here, we only use transform (not fit_transform) on X_test.
# This ensures that the test data is scaled using the same mean and standard deviation learned from the training set.
# ✅ This prevents data leakage (you should never learn scaling parameters from test data).



```

#What is **t-SNE**?

- t-SNE is a dimensionality reduction technique, mainly used for visualization.

- It reduces high-dimensional data (e.g., 100s or 1000s of features) down to 2 or 3 dimensions, so we can plot it easily.

- Unlike PCA (linear method), t-SNE is non-linear and focuses on preserving local structure → meaning, points that are close in high dimensions remain close in 2D/3D.

```python
#########Plotting TSNE plot to check whether problem is linear or not#######
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.manifold import TSNE
fig, ax = plt.subplots()
# Creates a new figure (fig) and a set of axes (ax) using Matplotlib.
# This sets up the plotting area.
m = TSNE(learning_rate=50)
# Creates a t-SNE model with learning_rate=50.
# learning_rate controls how t-SNE optimizes the embedding.
# (Common range: 10–200; smaller → more local detail, larger → more global structure).
# m is now a t-SNE object.

X_tsne = m.fit_transform(X)
# Applies t-SNE to  dataset X.
# fit_transform() reduces high-dimensional data into 2D coordinates.
# The result X_tsne is an array with two columns (for comp-1 and comp-2).

combined_PRAD["y"] = Y_train
# Adds a new column "y" to the combined_PRAD DataFrame, containing your training labels (Y_train).
# This will be used for coloring the scatter plot.

combined_PRAD["comp-1"] = X_tsne[:,0]
combined_PRAD["comp-2"] = X_tsne[:,1]
# Adds two new columns to the DataFrame:
# "comp-1" → first t-SNE dimension (x-axis).
# "comp-2" → second t-SNE dimension (y-axis).

sns.scatterplot(x="comp-1", y="comp-2", hue=combined_PRAD.y.tolist(),
                palette=sns.color_palette('husl', 2),
                data=combined_PRAD).set(title="Cancer data T-SNE projection")
# sns.scatterplot (Seaborn) plots a scatter plot.

# x="comp-1", y="comp-2" → use t-SNE reduced components for axes.
# hue=combined_PRAD.y.tolist() → colors the points based on class labels.
# palette=sns.color_palette('husl', 2) → selects a color palette with 2 distinct colors (since you have 2 classes: e.g., cancerous vs non-cancerous).
# data=combined_PRAD → uses your DataFrame as input.
# .set(title="Cancer data T-SNE projection") → sets the title of the plot.


plt.savefig("TSNE-plot.png", dpi = 600)
```

#**Confusion Matrix**
- A confusion matrix is a performance evaluation tool used in classification problems (machine learning, statistics, AI, etc.). It shows how well a model’s predictions match the actual outcomes by organizing results into a table.

- It compares the actual labels vs. the predicted labels and helps you see where the model is making correct predictions and where it is making mistakes.

#**Explanation:**

- **True Positive (TP):** Correctly predicted positive cases.

- **True Negative (TN):** Correctly predicted negative cases.

- **False Positive (FP):** Model predicted positive, but it was actually negative (Type I error).

- **False Negative (FN):** Model predicted negative, but it was actually positive (Type II error).

**From the confusion matrix, we can calculate:**

- **Accuracy:** (TP + TN) / Total

- **Precision:** TP / (TP + FP)

- **Recall (Sensitivity):** TP / (TP + FN)

- **Specificity:** TN / (TN + FP)

- **F1 Score:** Harmonic mean of Precision and Recall

#**👉 Example:**
- If we are predicting whether a patient has anemia or not:

- **TP:** Patients with anemia correctly identified as anemic.

- **TN:** Healthy patients correctly identified as not anemic.

- **FP:** Healthy patients wrongly predicted as anemic.

- **FN:** Anemic patients wrongly predicted as healthy.

```python
# Assuming you have trained your KNN model and have X_test and Y_test
# KNN is your trained K Nearest Neighbors model
# Get predictions from the KNN model
Y_pred = KNN.predict(X_test)

# Create the confusion matrix

cm = confusion_matrix(Y_test, Y_pred)
# Computes the confusion matrix by comparing:
# Y_test = true labels
# Y_pred = predicted labels (from your model).
# The result cm is a 2D array showing counts:

# Display the confusion matrix using ConfusionMatrixDisplay
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=KNN.classes_)  # KNN.classes_ contains  class labels
# ConfusionMatrixDisplay is a utility that plots the confusion matrix.
# display_labels=KNN.classes_ ensures the class names are shown on the axes (instead of just 0/1).
disp.plot(cmap=plt.cm.Greens)
plt.xlabel('Predicted label', color='black')
plt.ylabel('True label', color='black')
plt.gca().tick_params(axis='x', colors='white')  # Modify ticks on x-axis
plt.gca().tick_params(axis='y', colors='white')  # Modify ticks on y-axis
plt.gcf().set_size_inches(10, 6)
plt.savefig("KNN.png", dpi=600)
plt.show()

```

#**ROC Curve (Receiver Operating Characteristic Curve)**
The ROC curve is a graphical representation used to evaluate the performance of a binary classifier (like Logistic Regression, SVM, Decision Tree, etc.).

**It shows the trade-off between:**

- True Positive Rate (TPR) / Sensitivity / Recall
- False Positive Rate (FPR)

**How the ROC curve is plotted:**

- X-axis: False Positive Rate (FPR)

- Y-axis: True Positive Rate (TPR)

- Each point on the curve represents a different classification threshold.

```python
# roc curve and auc
from sklearn.metrics import roc_curve
from sklearn.metrics import roc_auc_score
from matplotlib import pyplot
# predict_proba() gives the probability estimates for each class instead of hard labels (0/1).
KNN_probs = KNN.predict_proba(X_test)
KNN_probs = KNN_probs[:, 1]
KNN_auc = roc_auc_score(Y_test, KNN_probs)
KNN_fpr, KNN_tpr, _ = roc_curve(Y_test, KNN_probs)
# plot the roc curve for the model
fig = pyplot.figure(figsize=(7, 5))
pyplot.plot(KNN_fpr, KNN_tpr ,label='KNN =%.3f' % (KNN_auc))
params = {'legend.fontsize': 10,
          'legend.handlelength': 2}
pyplot.rcParams.update(params)
pyplot.legend()
pyplot.ylabel('Precision', fontsize=15)
pyplot.xlabel('Recall', fontsize=15)
pyplot.xticks(fontsize=12)
pyplot.yticks(fontsize=12)
#Show legend
pyplot.legend() #
#plt.savefig("PR_curve.png", dpi = 600)
pyplot.show()
```

#**Linear Data**

- Data is linearly separable if you can draw a straight line (in 2D), or a flat plane (in 3D), or a hyperplane (in higher dimensions) to separate the classes.

- **Example:** Imagine classifying points where all red points are on one side and all blue points are on the other side of a straight line.

- **Simple rule:** If a straight line can separate the classes → Linear Data.

#**Non-Linear Data**

- Data is non-linearly separable if you cannot separate it with a straight line.

- We need a curved boundary (circle, parabola, wave, etc.) to separate the classes.

- **Example:** Blue points inside a circle and red points outside. A straight line won’t work; you need a curved line.

- **Simple rule:** If you need a curve to separate the classes → Non-Linear Data.

**Quick Analogy:**

- **Linear Data:** Like cutting a cake with one straight knife cut.

- **Non-Linear Data:** Like cutting around the cream filling of a donut — you need a curve.

#Support Vector Machine (**SVM**) – Linear Model
- An SVM (Support Vector Machine) is a supervised machine learning algorithm that can be used for classification and regression.

- A Linear SVM model is the simplest form of SVM, where the algorithm tries to find a straight line (in 2D) or a hyperplane (in higher dimensions) that best separates the data into different classes.

#**Advantages of Linear SVM:**

- Works well when data is linearly separable.

- Simple, efficient, and interpretable.

- Finds the optimal boundary with maximum margin.

#**Limitations:**

- If data is not linearly separable, performance drops.

- For that, we use Non-Linear SVM with kernels (e.g., RBF, polynomial).

```python
#SVC_linear
from sklearn.svm import SVC
svm_linear = SVC(kernel='linear', probability=True, random_state=40)
svm_linear.fit(X_train,Y_train).decision_function(X_test)
prediction = svm_linear.predict(X_test)
Accuracy = accuracy_score(Y_test,prediction)
print('accuracy',Accuracy)
#Printing precision, recall and f1_scores
print("Precision value: ", round(precision_score(Y_test,prediction),2) )
print("Recall_score: ", recall_score(Y_test,prediction))
print("F1_score: ", f1_score(Y_test,prediction))
```

```python
# Assuming you have trained your SVM linear model and have X_test and Y_test
# svm_linear is your trained Support Vector Machine with linear kernel model

# Get predictions from the SVM linear model
Y_pred_svm_linear = svm_linear.predict(X_test)

# Create the confusion matrix
cm_svm_linear = confusion_matrix(Y_test, Y_pred_svm_linear)

# Display the confusion matrix using ConfusionMatrixDisplay
disp_svm_linear = ConfusionMatrixDisplay(confusion_matrix=cm_svm_linear, display_labels=svm_linear.classes_)
disp_svm_linear.plot(cmap=plt.cm.Greens)
plt.xlabel('Predicted label', color='black')
plt.ylabel('True label', color='black')
plt.gca().tick_params(axis='x', colors='white')  # Modify ticks on x-axis
plt.gca().tick_params(axis='y', colors='white')  # Modify ticks on y-axis
plt.gcf().set_size_inches(10, 6)
plt.savefig("svm_linear.png", dpi=600)
plt.show()
svm_linear_probs = svm_linear.predict_proba(X_test)
svm_linear_probs = svm_linear_probs[:, 1]
svm_linear_auc = roc_auc_score(Y_test, svm_linear_probs)
svm_linear_fpr, svm_linear_tpr, _ = roc_curve(Y_test, svm_linear_probs)
# plot the roc curve for the model
fig = pyplot.figure(figsize=(7, 5))
pyplot.plot(svm_linear_fpr, svm_linear_tpr ,label='SVM_Linear =%.3f' % (svm_linear_auc))
params = {'legend.fontsize': 10,
          'legend.handlelength': 2}
pyplot.rcParams.update(params)
pyplot.legend()
pyplot.ylabel('True Positive Rate', fontsize=15)
pyplot.xlabel('False Positive Rate', fontsize=15)
pyplot.xticks(fontsize=12)
pyplot.yticks(fontsize=12)
#Show legend
pyplot.legend() #
#plt.savefig("AUC_ROC.png", dpi = 600)
pyplot.show()
```

#**SVC with Polynomial Kernel**

- When the data is not linearly separable (can’t be separated by a straight line or hyperplane), we can use the kernel trick.

- A Polynomial Kernel allows SVM to classify data that has non-linear relationships by mapping it into a higher-dimensional feature space.

#**GaussianNB**

- GaussianNB is a Naive Bayes classifier from sklearn.naive_bayes.

- Naive Bayes is a simple probabilistic machine learning algorithm based on Bayes’ Theorem.

- It’s called “naive” because it assumes that all features are independent (which is often not true in real-world data, but works surprisingly well).

- GaussianNB is a specific version of Naive Bayes used when the features are continuous and assumed to follow a normal (Gaussian) distribution.

- **Real-life analogy:**

- Imagine you’re a doctor trying to predict if a patient has a disease.

- You measure continuous values like blood pressure, hemoglobin, or temperature.

- For each disease (class), you assume these values follow a bell-shaped (Gaussian) curve.

- If the patient’s values are closer to the mean of "Disease A" distribution, you classify them as having Disease A.

#**MLPClassifier (Multi-Layer Perceptron Classifier)**

#**AdaBoostClassifier (Adaptive Boosting Classifier)**

- AdaBoostClassifier is an ensemble learning method in scikit-learn.

- It combines many weak learners (usually shallow decision trees called stumps) into one strong classifier.

- "Adaptive" means it focuses more on the mistakes made by previous learners when building the next one.

#**How it works step by step:**

- Train a weak learner (e.g., decision stump) on the data.

- Check which samples were misclassified.

- Assign higher weights to those misclassified samples.

- Train the next weak learner, paying more attention to hard-to-classify cases.

- Repeat for several iterations, then combine all learners’ votes (weighted majority voting).

- The final model is much stronger than any single weak learner.

```python
from sklearn.ensemble import AdaBoostClassifier
# define the model
AB = AdaBoostClassifier()
AB.fit(X_train, Y_train)
prediction = AB.predict(X_test)
Accuracy = accuracy_score(Y_test,prediction)
print('accuracy',Accuracy)
#Printing precision, recall and f1_scores
print(precision_score(Y_test,prediction))
print(recall_score(Y_test,prediction))
print(f1_score(Y_test,prediction))
# Assuming you have trained your SVM poly model and have X_test and Y_test
# svm_poly is your trained SVM with polynomial kernel model

# Get predictions from the SVM poly model
Y_pred_AB = AB.predict(X_test)

# Create the confusion matrix
cm_AB = confusion_matrix(Y_test, Y_pred_AB)

# Display the confusion matrix using ConfusionMatrixDisplay
disp_AB = ConfusionMatrixDisplay(confusion_matrix=cm_AB, display_labels=AB.classes_)
disp_AB.plot(cmap=plt.cm.Greens)
plt.xlabel('Predicted label', color='black')
plt.ylabel('True label', color='black')
plt.gca().tick_params(axis='x', colors='white')  # Modify ticks on x-axis
plt.gca().tick_params(axis='y', colors='white')  # Modify ticks on y-axis
plt.gcf().set_size_inches(10, 6)
plt.savefig("AB.png", dpi=600)
plt.show()


AB_probs = AB.predict_proba(X_test)
AB_probs = AB_probs[:, 1]
AB_auc = roc_auc_score(Y_test, AB_probs)
AB_fpr, AB_tpr, _ = roc_curve(Y_test, AB_probs)
# plot the roc curve for the model
fig = pyplot.figure(figsize=(7, 5))
pyplot.plot(AB_fpr, AB_tpr ,label='AB =%.3f' % (AB_auc))
params = {'legend.fontsize': 10,
          'legend.handlelength': 2}
pyplot.rcParams.update(params)
pyplot.legend()
pyplot.ylabel('True Positive Rate', fontsize=15)
pyplot.xlabel('False Positive Rate', fontsize=15)
pyplot.xticks(fontsize=12)
pyplot.yticks(fontsize=12)
#Show legend
pyplot.legend() #
#plt.savefig("AUC_ROC.png", dpi = 600)
pyplot.show()
```

#**RandomForestClassifier (Scikit-learn)**

- RandomForestClassifier is an ensemble learning method that builds a “forest” of decision trees.

- Each tree is trained on a slightly different subset of the data + features (using bagging and feature randomness).

- The final prediction is made by majority vote of all the trees.

**How it works step by step:**

- Take multiple random samples (with replacement) from the dataset → bootstrap sampling.

- Train a decision tree on each sample, but at each split, the tree only considers a random subset of features (not all).

- Repeat for many trees (e.g., 100).

- For classification → each tree votes for a class, and the majority wins.

**Advantages:**

- Handles large datasets well.

- Robust to noise and outliers.

- Less prone to overfitting than a single decision tree.

- Works for both classification and regression.

**Disadvantages:**

- Slower to train compared to a single tree.

- Harder to interpret (a “black box” compared to one decision tree).

#**Real-Life Analogy: Asking a Group of Doctors**

**Imagine you want a diagnosis:**

- If you ask just one doctor (decision tree), their opinion may be biased.

- If you ask 100 doctors (random forest), each looking at slightly different aspects of your medical history, then take the majority vote, you’re more likely to get the correct diagnosis.

- That’s Random Forest — many weak decision-makers combine to form a strong, reliable decision.

```python
from sklearn.ensemble import RandomForestClassifier
# define the model
RF = RandomForestClassifier()
RF.fit(X_train, Y_train)
prediction = RF.predict(X_test)
Accuracy = accuracy_score(Y_test,prediction)
print('accuracy',Accuracy)
#Printing precision, recall and f1_scores
print(precision_score(Y_test,prediction))
print(recall_score(Y_test,prediction))
print(f1_score(Y_test,prediction))
# Assuming you have trained your SVM poly model and have X_test and Y_test
# svm_poly is your trained SVM with polynomial kernel model

# Get predictions from the SVM poly model
Y_pred_RF = RF.predict(X_test)

# Create the confusion matrix
cm_RF = confusion_matrix(Y_test, Y_pred_RF)

# Display the confusion matrix using ConfusionMatrixDisplay
disp_RF = ConfusionMatrixDisplay(confusion_matrix=cm_RF, display_labels=RF.classes_)
disp_RF.plot(cmap=plt.cm.Greens)
plt.xlabel('Predicted label', color='black')
plt.ylabel('True label', color='black')
plt.gca().tick_params(axis='x', colors='white')  # Modify ticks on x-axis
plt.gca().tick_params(axis='y', colors='white')  # Modify ticks on y-axis
plt.gcf().set_size_inches(10, 6)
plt.savefig("RF.png", dpi=600)
plt.show()


RF_probs = RF.predict_proba(X_test)
RF_probs = RF_probs[:,1]
RF_auc = roc_auc_score(Y_test, RF_probs)
RF_fpr, RF_tpr, _ = roc_curve(Y_test, RF_probs)

# plot the roc curve for the model
fig = pyplot.figure(figsize=(7, 5))
pyplot.plot(RF_fpr, RF_tpr ,label='RF =%.3f' % (RF_auc))
params = {'legend.fontsize': 10,
          'legend.handlelength': 2}
pyplot.rcParams.update(params)
pyplot.legend()
pyplot.ylabel('True Positive Rate', fontsize=15)
pyplot.xlabel('False Positive Rate', fontsize=15)
pyplot.xticks(fontsize=12)
pyplot.yticks(fontsize=12)
#Show legend
pyplot.legend() #
#plt.savefig("AUC_ROC.png", dpi = 600)
pyplot.show()
```
